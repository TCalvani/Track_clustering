```python
import re

def extract_playlist_id(playlist_link):
    match = re.search(r'playlist/([a-zA-Z0-9]+)\?', playlist_link)
    if match:
        return match.group(1)
    else:
        return None

def extract_playlist_ids(playlist_links):
    playlist_ids = [extract_playlist_id(link) for link in playlist_links]
    return [pid for pid in playlist_ids if pid is not None]
```

```python

from datetime import datetime


def retrieve_all_tracks(playlist_id):
    try:
        results = sp.playlist_tracks(playlist_id)
        tracks = results['items']

        while results['next']:
            results = sp.next(results)
            tracks.extend(results['items'])

        return tracks
    except Exception as e:
        print(f"Error retrieving tracks for playlist {playlist_id}: {e}")
        return None

def get_track_genres(track_id):
    try:
        track_info = sp.track(track_id)
        time.sleep(1)
        artist_ids = [artist['id'] for artist in track_info['artists']]
        artists_info = sp.artists(artist_ids)
        genres = set()

        for artist_info in artists_info['artists']:
            genres.update(artist_info['genres'])

        return list(genres)
    except Exception as e:
        print(f"Error retrieving genres for track {track_id}: {e}")
        return list(genres)

def retrieve_track_features(track_ids):
    data = []
    print("Processing new playlist")

    for i in range(0, len(track_ids), 50):  # Batch processing (50 tracks per request)
        batch_track_ids = track_ids[i:i+50]

        try:
            tracks_info = sp.tracks(batch_track_ids)['tracks']

            for track_info in tracks_info:
                features = sp.audio_features(track_info['id'])[0]
                data.append({
                    'track_id': track_info['id'],
                    'track_name': track_info['name'],
                    'artist_name': track_info['artists'][0]['name'],
                    'album_name': track_info['album']['name'],
                    'popularity': track_info['popularity'],
                    **features  
                })
                r = random.randint(1, 4)
                time.sleep(r)

        except Exception as e:
            print(f"Error processing batch of tracks: {e}")

    return data


def create_dataframe(playlists):
    data = []
    for playlist_id in playlists:
        start=datetime.now()
        playlist_tracks = retrieve_all_tracks(playlist_id)

        if playlist_tracks is not None:
            track_ids = [track['track']['id'] for track in playlist_tracks]
            playlist_features = retrieve_track_features(track_ids)
            data.extend(playlist_features)

        else:
            continue
        print("Time to complete: ",datetime.now()-start)

        time.sleep(5)

    data = [item for item in data if item is not None]

    df = pd.DataFrame(data)

    return df
 

try:
    for p in pls:
        genre=p[0]
        playlist_ids=extract_playlist_ids(p[1:])
        df = create_dataframe(playlist_ids)
        df['genre']=genre
        tracks_df = tracks_df.merge(df, how='outer')
except Exception as e:
    print(f"Error creating DataFrame: {e}")
```
